# 패키지

```python
from torchvision import transforms
from PIL import Image
import os
```

# 이미지 경로

```python
img = Image.open('C:/Users/양준영/pytorch/train/냥이.jpg')
print(img.size)
```

# 이미지 출력

```python
img
```

# RESIZE

```python
resize = transforms.Compose([
    transforms.Resize((244,244))
])

resize(img)
```

# flip 상하

```python
flip = transforms.Compose([
    transforms.RandomVerticalFlip(p = 1)
])

flip(img)
```

# flip 좌우

```python
flip = transforms.Compose([
    transforms.RandomHorizontalFlip(p = 0.5) #일정확률로 좌우를 변경할지 정하는 것이 가능함.
])

flip(img)
```

# 랜덤 회전

```python
Random_rotation = transforms.Compose([
    transforms.RandomRotation(degrees=30) #랜덤으로 30도 회전
])

Random_rotation(img)
```

# 랜덤 색상 변경

```python

Random_colorJitter = transforms.Compose([
    transforms.ColorJitter(brightness = 0.9,contrast = 0.2,saturation = 0.2,hue=0.1) # 밝기 ,대비, 채도 , 색조
])

Random_colorJitter(img)
```

# 랜덤 자르기

```python
RandomResizedCrop = transforms.Compose([
    transforms.RandomResizedCrop(244)
])

RandomResizedCrop(img)
```

# 특정 영역 자르기

```python
Crop = transforms.Compose([
    transforms.CenterCrop((img.height // 2, img.width)),  # 이미지의 상단 절반 유지
    transforms.Resize((224, 224)),
])

Crop(img)
```

# 가우시안 블러(이미지를 흐릿하게 만들어 패턴이 덜보이게함.)

```python
GaussianBlur = transforms.Compose([
    transforms.GaussianBlur(kernel_size=(5,5))
])

GaussianBlur(img)
```

# Nomalize

```python
Nomalize = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

Nomalize(img)
```
